```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_close_prices = pd.read_csv('./close_prices.csv')
data_close_prices = data_close_prices.drop('date', axis = 1)
data_close_prices.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 10)
pca.fit(data_close_prices)
```

```python
[pca.explained_variance_ratio_[:i].sum() for i in range(len(pca.explained_variance_))]
```

```python
X = pca.transform(data_close_prices)
X = [i[0] for i in X]
X
```

```python
data_djia = pd.read_csv('./djia_index.csv')
data_djia.head()
```

```python
np.corrcoef(X, data_djia['^DJI'])
```
